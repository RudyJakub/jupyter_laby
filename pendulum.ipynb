```python
import matplotlib.pyplot as plt
plt.title("Zależność okresu wahadła od długości nici", loc = 'left')
plt.xlabel("Długość nici l [cm]")
plt.ylabel("Okres T [s]")

plt.grid()
plt.scatter(sorted(lengths2), sorted(w2))
plt.legend(['T(l)']) 
```

```python

w3 = [ww ** 2 for ww in w2]
plt.title("Zależność kwadratu okresu wahadła od długości nici", loc = 'left')
plt.xlabel("Długość nici l [cm]")
plt.ylabel("Kwadrat okresu T^2 [s^2]")

plt.grid()
plt.scatter(sorted(lengths2), sorted(w3))
plt.legend(['T^2(l)']) 
```

```python
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(lengths2, w3)
print(slope, intercept, r, p, std_err)
```
